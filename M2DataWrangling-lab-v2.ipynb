### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# 1. Show all unique values in the Country column
unique_countries = df['Country'].unique()
print("Unique values in Country column:")
print(unique_countries)

# 2. Count occurrences of each value (helps spot typos)
country_counts = df['Country'].value_counts(dropna=False)
print("\nCountry value counts:")
print(country_counts)

# 3. Identify potential inconsistencies: e.g., entries that are very short or numeric
inconsistent_entries = df[df['Country'].str.len() < 2]  # short entries
print("\nPotential inconsistent entries (very short names):")
print(inconsistent_entries[['Country']])

numeric_entries = df[df['Country'].str.isnumeric()]  # purely numeric entries
print("\nPotential numeric entries:")
print(numeric_entries[['Country']])
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
country_mapping = {
    "USA": "United States",
    "U.S.A": "United States",
    "US": "United States",
    "United States of America": "United States",
    "UK": "United Kingdom",
    "U.K.": "United Kingdom",
    "England": "United Kingdom",
    # Add more mappings as needed
}

df['Country'] = df['Country'].replace(country_mapping)

# ===============================
# Standardize 'EdLevel' column
# ===============================
edlevel_mapping = {
    "Bachelor’s degree": "Bachelor",
    "Bachelors degree": "Bachelor",
    "Master’s degree": "Master",
    "Masters degree": "Master",
    "Some college/university study without earning a degree": "Some College",
    "Secondary school": "High School",
    # Add more mappings as needed
}

df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

# ===============================
# Verify changes
# ===============================
print(df['Country'].value_counts())
print(df['EdLevel'].value_counts())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
# Count missing values per column
missing_counts = df.isna().sum()

# Sort columns by number of missing values (descending)
missing_sorted = missing_counts.sort_values(ascending=False)

print("Columns with the highest number of missing values:")
print(missing_sorted)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
# Option 1: Impute with mean
# ================================
mean_value = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly_mean_imputed'] = df['ConvertedCompYearly'].fillna(mean_value)

# ================================
# Option 2: Impute with median
# ================================
median_value = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly_median_imputed'] = df['ConvertedCompYearly'].fillna(median_value)

# ================================
# Verify imputation
# ================================
print("Missing values after mean imputation:", df['ConvertedCompYearly_mean_imputed'].isna().sum())
print("Missing values after median imputation:", df['ConvertedCompYearly_median_imputed'].isna().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
# Select categorical columns (object type)
cat_cols = df.select_dtypes(include='object').columns

# Impute missing values with the most frequent value (mode) for each categorical column
for col in cat_cols:
    most_frequent = df[col].mode()[0]  # get the mode
    df[col] = df[col].fillna(most_frequent)

# Verify that there are no missing values in categorical columns
print(df[cat_cols].isna().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>
